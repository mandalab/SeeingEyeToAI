# 02 Visualize Averaged Fixation Heatmaps
 
Load and display the precomputed heatmaps to sanity-check the smoothing & averaging pipeline.

```python
import pickle
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
HEATMAP_DIR = Path('./pickles/eyetracking_data')
pkl_files = sorted(HEATMAP_DIR.glob('*_fixation_gBr.pkl'))
print("Found heatmap pickles:", [p.name for p in pkl_files])
```

### Load a heatmap dictionary

```python
with open(pkl_files[0], 'rb') as f:
    heatmaps = pickle.load(f)

print(f"Loaded heatmaps for {len(heatmaps)} videos, each with {len(next(iter(heatmaps.values())))} frames")
```

### Display first frame of first 3 videos

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, (vid, frames) in zip(axes, list(heatmaps.items())[:3]):
    ax.imshow(frames[0], cmap='jet')
    ax.set_title(vid)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

### Display all frames for a single video

```python
sample_vid, sample_frames = next(iter(heatmaps.items()))
print(f"Showing all {len(sample_frames)} frames for video: {sample_vid}")
fig, axes = plt.subplots(1, len(sample_frames), figsize=(len(sample_frames)*3, 3))
for ax, frame in zip(axes, sample_frames):
    ax.imshow(frame, cmap='jet')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
